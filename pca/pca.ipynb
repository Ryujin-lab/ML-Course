```python
from sklearn import datasets
from sklearn.model_selection import train_test_split
```

```python
iris = datasets.load_iris()
atribute = iris.data
label = iris.target

xTrain, xTest, yTrain, yTest = train_test_split(atribute, label, test_size = 0.2)
```

```python
from sklearn import tree
decicionTree = tree.DecisionTreeClassifier()
model1 = decicionTree.fit(xTrain,yTrain)
model1.score(xTest, yTest)
```

```python
#pca 4
from sklearn.decomposition import PCA

pca = PCA(n_components = 4)

pcaAtribuete = pca.fit_transform(xTrain)
pca.explained_variance_ratio_
```

```python
#pca 2
from sklearn.decomposition import PCA

pca = PCA(n_components = 2)

pcaAtribuete = pca.fit_transform(xTrain)
pca.explained_variance_ratio_
```

```python
xTrainPCA  = pca.fit_transform(xTrain)
xTestPCA = pca.fit_transform(xTest)
```

```python
model2 = decicionTree.fit(xTrainPCA,yTrain)
model2.score(xTestPCA,yTest)
```
