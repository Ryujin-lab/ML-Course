```python
from sklearn.ensemble import ExtraTreesClassifier

X = new_data.iloc[:,0:24]  #independent columns
y = new_data.iloc[:,-1]    #target column i.e price range
model = ExtraTreesClassifier()

model.fit(X,y)
print(model.feature_importances_) #use inbuilt class feature_importances of tree based classifiers
#plot graph of feature importances for better visualization
plt.figure(figsize=(10,5))
feat_importances = pd.Series(model.feature_importances_, index=X.columns)
feat_importances.nlargest(10).plot(kind='barh')
plt.show()
```

```python
# "homogenity", sm, energ
# 
# y
feature = [
    # 'dissimilarity_0',  'dissimilarity_45', 'dissimilarity_90', 'dissimilarity_135', 
    # 'correlation_0',    'correlation_45',   'correlation_90',   'correlation_135', 
    'homogeneity_0',    'homogeneity_45',   'homogeneity_90',   'homogeneity_135', 
    # 'contrast_0',       'contrast_45',      'contrast_90',      'contrast_135', 
    'ASM_0',            'ASM_45',           'ASM_90',           'ASM_135', 
    'energy_0',         'energy_45',        'energy_90',        'energy_135']

feature90 = ['dissimilarity_90',
'correlation_90',
'homogeneity_90',
'contrast_90',
'ASM_90',
'energy_90']
```

```python
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsClassifier
from sklearn import metrics
from sklearn.preprocessing import MinMaxScaler
from sklearn.tree import DecisionTreeClassifier
from scipy import stats
```

```python

x = new_data[feature]
y = new_data["label"]

scaler = MinMaxScaler()
scaler.fit(x)
scaled_x = scaler.transform(x)
scaled_x = pd.DataFrame(scaled_x, columns= x.columns)

x_train, x_test, y_train, y_test = train_test_split(scaled_x, y, test_size=0.2, random_state=22)
```

```python
scores = {}
s = []
for k in range (1,31) :
    knn = KNeighborsClassifier(k)
    knn.fit(x_train,y_train)
    y_pred = knn.predict(x_test)
    acc = metrics.accuracy_score(y_test, y_pred)
    scores[f"k({k})"] = acc
    s.append(acc)
np.average(s)
```

```python
from sklearn.decomposition import PCA
 
pca = PCA(n_components=6)
pca_attributes = pca.fit_transform(scaled_x)
 
pca.explained_variance_ratio_
# melihat variance dari setiap atribut/ pake 3 aja
```

```python
pca = PCA(n_components = 4)
X_train_pca = pca.fit_transform(x_train)
X_test_pca = pca.fit_transform(x_test)

scores_pca = {}
s_pca = []
for k in range (1,31) :
    knn = KNeighborsClassifier(k)
    knn.fit(X_train_pca,y_train)
    # y_pred = knn.predict(X_test_pca)
    acc = knn.score(X_test_pca, y_test)
    scores_pca[f"k({k})"] = acc
    s_pca.append(acc)
np.average(s_pca)
```

```python
clean_data = new_data.copy()
clean_data.info()
```

```python
z_scores = stats.zscore(clean_data)
abs_z_scores = np.abs(z_scores)
filtered_entries = (abs_z_scores < 3).all(axis=1)
clean_data_outer = clean_data[filtered_entries]
clean_data_outer.info()
```

```python
x_clear = clean_data_outer[feature]
y_clear = clean_data_outer["label"]

scaler_clear = MinMaxScaler()
scaler_clear.fit(x_clear)
scaled_x_clear = scaler_clear.transform(x_clear)
scaled_x_clear = pd.DataFrame(scaled_x_clear, columns= x_clear.columns)

x_clear_train, x_clear_test, y_clear_train, y_clear_test = train_test_split(scaled_x_clear, y_clear, test_size=0.2, random_state=22)
```

```python
scores_clear = {}
for k in range (1,31) :
    knn = KNeighborsClassifier(k)
    knn.fit(x_clear_train,y_clear_train)
    y_clean_pred = knn.predict(x_clear_test)
    scores_clear[f"k({k})"] = metrics.accuracy_score(y_clear_test, y_clean_pred)
scores_clear
```
